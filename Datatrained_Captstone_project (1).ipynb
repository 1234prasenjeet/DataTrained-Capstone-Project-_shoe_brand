```python
# Function to extract data for Table1
def extract_data_table1(soup):
    # Example: Extracting shoe name, category, number of colors, and price
    shoe_name_element = soup.find('h1', class_='product-title')
    category_element = soup.find('span', class_='product-category')
    colors_elements = soup.find_all('div', class_='color-selector-item')
    price_element = soup.find('span', class_='product-price')

    # Check if elements exist before extracting text
    shoe_name = shoe_name_element.text.strip() if shoe_name_element else ''
    category = category_element.text.strip() if category_element else ''
    num_colors = len(colors_elements) if colors_elements else 0
    price = price_element.text.strip() if price_element else ''

    # Return data as a list
    return [[shoe_name, category, num_colors, price]]
```

```python
# Function to extract data for Table2
def extract_data_table2(soup):
    # Example: Extracting count of sizes, colors, and style code
    count_sizes_elements = soup.find_all('div', class_='size-selector-item')
    colors_elements = soup.find_all('div', class_='color-selector-item')
    style_code_element = soup.find('span', class_='product-style-code')

    # Check if elements exist before extracting text
    count_sizes = len(count_sizes_elements) if count_sizes_elements else 0
    colors = [color['data-color'] for color in colors_elements] if colors_elements else []
    style_code = style_code_element.text.strip() if style_code_element else ''

    # Return data as a list
    return [[count_sizes] + colors + [style_code]]
```

```python
# Function to extract data for Table3
def extract_data_table3(soup):
    # Example: Extracting reviews, size, comfort, durability, star rating
    reviews_element = soup.find('span', class_='product-reviews-count')
    size_element = soup.find('span', class_='size-selector-item-selected')
    comfort_element = soup.find('span', class_='product-feature-comfort')
    durability_element = soup.find('span', class_='product-feature-durability')
    star_rating_element = soup.find('span', class_='product-rating-stars')

    # Check if elements exist before extracting text
    reviews = reviews_element.text.strip() if reviews_element else ''
    size = size_element.text.strip() if size_element else ''
    comfort = comfort_element.text.strip() if comfort_element else ''
    durability = durability_element.text.strip() if durability_element else ''
    star_rating = star_rating_element.text.strip() if star_rating_element else ''

    # Return data as a list
    return [[reviews, size, comfort, durability, star_rating]]
```
